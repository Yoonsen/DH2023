### Bygg korpus fra innholdsord

```python
dh.Corpus(doctype='digibok', fulltext='demokrati', from_year=1990, to_year=2000, limit=5)
```

```python
dh.Corpus(doctype='digibok', fulltext='her OR kan OR du OR legge OR inn OR flere OR søkeord', from_year=1960, to_year=1980, limit=5)
```

### Legg korpuset i en variabel

Variabelen brukes til å referere til korpuset senere

```python
korpus = dh.Corpus(ddk="9%", subject="reindrift", from_year=1960, to_year=2020, limit=5)
korpus
```
